## Bringing Features Onto the Same Scale: Feature Scaling
**Feature scaling** is a crucial preprocessing step for machine learning algorithms like **gradient descent** and **k-nearest neighbors (KNN)** to ensure features with larger ranges don't dominate the objective function. While **decision trees** and **random forests** are scale-invariant, most algorithms require scaling techniques like **normalization** and **standardization** for optimal performance.



---

```python
import pandas as pd
from sklearn.model_selection import train_test_split

df_wine = pd.read_csv('https://archive.ics.uci.edu/'
                       'ml/machine-learning-databases/'
                       'wine/wine.data', header=None)

df_wine.columns = ['Class label', 'Alcohol',
                    'Malic acid', 'Ash',
                    'Alcalinity of ash', 'Magnesium',
                    'Total phenols', 'Flavanoids',
                    'Nonflavanoid phenols',
                    'Proanthocyanins',
                    'Color intensity', 'Hue',
                    'OD280/OD315 of diluted wines',
                    'Proline']

X, y = df_wine.iloc[:, 1:].values, df_wine.iloc[:, 0].values

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3,
                                                    random_state=0,
                                                    stratify=y)
```

### 1. Normalization (Min-Max Scaling)
Normalization typically refers to rescaling features to a bounded interval, most commonly $[0, 1]$. It is highly useful when we need values to exist strictly within a limited range.

*   **Equation:** $$x^{(i)}_{norm} = \frac{x^{(i)} - x_{min}}{x_{max} - x_{min}}$$
*   Here, $x^{(i)}$ is a particular example, $x_{min}$ is the smallest value in a feature column, and $x_{max}$ is the largest value.

```python
from sklearn.preprocessing import MinMaxScaler

mms = MinMaxScaler()

X_train_norm = mms.fit_transform(X_train)
X_test_norm  = mms.transform(X_test)
```

### 2. Standardization
Standardization centers feature columns at a mean of $0$ with a standard deviation of $1$, giving them the properties of a standard normal distribution.

*   This approach is widely practical for linear models (like logistic regression and SVMs) and optimization algorithms, as it makes learning weights easier.
*   Unlike normalization, standardization does not bound the data to a specific range, meaning it maintains useful information about outliers and makes algorithms less sensitive to them[cite: 1].
*   **Equation:** $$x^{(i)}_{std} = \frac{x^{(i)} - \mu_x}{\sigma_x}$$
*   Here, $\mu_x$ is the sample mean of a particular feature column, and $\sigma_x$ is the corresponding standard deviation.

**Manual Implementation:**

```python
import numpy as np

ex = np.array([0, 1, 2, 3, 4, 5])

standardized = (ex - ex.mean()) / ex.std()
print('standardized:', standardized)

normalized = (ex - ex.min()) / (ex.max() - ex.min())
print('normalized:', normalized)
```

**Scikit-Learn Implementation:**

```python
from sklearn.preprocessing import StandardScaler

stdsc = StandardScaler()

X_train_std = stdsc.fit_transform(X_train)
X_test_std  = stdsc.transform(X_test)
```

*Note: It is crucial to fit the `StandardScaler` class only once on the training data, and use those established parameters to transform the test dataset or any new data points*.

### 3. Robust Scaling
Other advanced methods are available for specific dataset challenges.

*   **Handling Outliers:** For small datasets containing many outliers, or when a model is prone to **overfitting**, the `RobustScaler` is recommended.
*   **Mechanism:** It operates independently on each feature column by removing the median value and scaling the data according to the 1st (25th percentile) and 3rd (75th percentile) quartiles.
*   **Result:** This ensures extreme values and outliers become significantly less pronounced in the dataset.
