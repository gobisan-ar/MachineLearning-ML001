## **Splitting a Dataset into Separate Training and Test Datasets**

Splitting data into training and testing sets is crucial for machine learning. Comparing a model's predictions to true labels in a dedicated test set provides an **unbiased performance evaluation**, giving you a realistic idea of how the model will perform in the real world before deployment.

---

### The Wine Dataset

To explore data preprocessing and feature selection techniques for dimensionality reduction, we will use the **Wine dataset**.

*   **Source:** An open-source dataset from the [UCI Machine Learning Repository](https://archive.ics.uci.edu/ml/datasets/Wine).
*   **Size:** 178 wine examples.
*   **Features:** 13 features describing their various chemical properties.

```python
import pandas as pd
import numpy as np
```

```python
df_wine = pd.read_csv('https://archive.ics.uci.edu/'
                       'ml/machine-learning-databases/'
                       'wine/wine.data', header=None)

df_wine.columns = ['Class label', 'Alcohol',
                    'Malic acid', 'Ash',
                    'Alcalinity of ash', 'Magnesium',
                    'Total phenols', 'Flavanoids',
                    'Nonflavanoid phenols',
                    'Proanthocyanins',
                    'Color intensity', 'Hue',
                    'OD280/OD315 of diluted wines',
                    'Proline']

print('Class labels', np.unique(df_wine['Class label']))

df_wine.head()
```

### **1. Using Scikit-Learn's train_test_split**

A convenient method to randomly partition a dataset is scikit-learn's `train_test_split` function from the `model_selection` submodule.

```python
from sklearn.model_selection import train_test_split

X, y = df_wine.iloc[:, 1:].values, df_wine.iloc[:, 0].values

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3,
                                                    random_state=0,
                                                    stratify=y)
```

### **2. Choosing an Appropriate Partition Ratio**
Selecting the right split ratio is fundamentally about balancing the learning algorithm's need for data against the need to accurately estimate the generalization error[cite: 1].

*   **Standard Splits:** In practice, the most commonly used dataset splits are 60:40, 70:30, or 80:20[cite: 1]. The choice often depends heavily on the initial dataset's size[cite: 1].
*   **Large Datasets:** For massive datasets (e.g., more than 100,000 training examples), splits like 90:10 or 99:1 are common and appropriate[cite: 1]. Withholding just 10,000 examples is often sufficient to get a good estimate of generalization performance[cite: 1].
*   **Stratification:** By passing the target array `y` to the `stratify` parameter, you guarantee that both the training and test datasets retain the same class proportions as the original dataset[cite: 1].
*   **Retraining Best Practice:** Rather than simply discarding the test data after evaluation, it is common practice to retrain your chosen classifier on the entire dataset to maximize its predictive capabilities[cite: 1].
