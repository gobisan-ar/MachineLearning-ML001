# Implementing a Multilayer Artificial Neural Network

## Introduction to Artificial Neural Networks (NNs)

Deep learning is a highly prominent subfield of machine learning that focuses on training artificial neural networks (NNs) with many layers.

* **Historical Context:** The foundational concepts of NNs are based on models of the human brain from the 1940s, primarily introduced by Warren McCulloch and Walter Pitts.
* **AI Winters:** Following the invention of Rosenblatt's perceptron in the 1950s, interest waned because there was no effective method for training multilayer networks.
* **The Resurgence:** Interest was renewed in 1986 when D.E. Rumelhart, G.E. Hinton, and R.J. Williams popularized the backpropagation algorithm, enabling efficient training of multilayer NNs.
* **Modern Applications:** Today, complex deep learning algorithms power state-of-the-art solutions for image and voice recognition, modeling virus mutations, predicting resource needs, and managing extreme weather events.

## Single-Layer Neural Network Recap

To understand multilayer architectures, it is helpful to review single-layer models like the **ADAptive LInear NEuron (Adaline)**.

* **Gradient Descent Optimization:** The Adaline algorithm uses gradient descent to learn the weight vector $\boldsymbol{w}$ and bias unit $b$ by minimizing a Mean Squared Error (MSE) loss function, denoted as $L(\boldsymbol{w})$.
* **Update Rules:** Weights and biases are updated in every epoch using the rule $\boldsymbol{w} := \boldsymbol{w} + \Delta\boldsymbol{w}$ and $b := b + \Delta b$, where the updates are scaled by a learning rate $\eta$. The partial derivative used for the weight update is defined as:

$$\frac{\partial L}{\partial w_j} = -\frac{2}{n} \sum_i (y^{(i)} - a^{(i)})x_j^{(i)}$$


* **Activation and Threshold:** Adaline uses a linear activation function $\sigma(z) = z = a$ to compute the gradient, followed by a threshold function to squash the continuous output into binary class predictions.
* **Stochastic Gradient Descent (SGD):** An alternative to standard gradient descent, SGD approximates the loss using a single sample or mini-batch. Its noisy nature helps models escape local minima, which is particularly beneficial for multilayer networks with nonlinear activation functions.

## Multilayer Perceptron (MLP) Architecture

Connecting multiple single neurons creates a feedforward neural network known as a **Multilayer Perceptron (MLP)**.

* **Structure:** An MLP typically consists of an input layer, one or more hidden layers, and an output layer. If a network contains more than one hidden layer, it is referred to as a **deep Neural Network (DNN)**.
* **Vanishing Gradient Problem:** Adding more layers introduces challenges such as vanishing gradients, which makes learning difficult and necessitates special deep learning algorithms.
* **Multiclass Classification:** To handle multiple classes, the output layer generalizes the one-versus-all (OvA) technique using one-hot vector representations.

## Activating a Neural Network via Forward Propagation

The process of training an MLP involves three primary steps: forward propagation, loss calculation, and backpropagation.

### Forward Propagation Steps

1. **Net Input Calculation:** For each unit in the hidden layer, the net input is the linear combination of inputs and weights.
2. **Nonlinear Activation:** A differentiable activation function $\sigma(\cdot)$ is applied to the net input. To solve complex problems, a nonlinear function like the **sigmoid (logistic) function** is typically used.
* The sigmoid function maps the net input onto a distribution in the continuous range of 0 to 1:

$$\sigma(z) = \frac{1}{1 + e^{-z}}$$





### Vectorized Implementation

To improve code efficiency and readability in Python (e.g., using NumPy), these calculations are vectorized to avoid computationally expensive `for` loops.

* **Hidden Layer Computations (for $n$ examples):**
* Net Input Matrix: $\boldsymbol{Z}^{(h)} = \boldsymbol{X}^{(in)}\boldsymbol{W}^{(h)T} + \boldsymbol{b}^{(h)}$
* Activation Matrix: $\boldsymbol{A}^{(h)} = \sigma(\boldsymbol{Z}^{(h)})$


* **Output Layer Computations:**
* Net Input Matrix: $\boldsymbol{Z}^{(out)} = \boldsymbol{A}^{(h)}\boldsymbol{W}^{(out)T} + \boldsymbol{b}^{(out)}$
* Final Output Activation: $\boldsymbol{A}^{(out)} = \sigma(\boldsymbol{Z}^{(out)})$
