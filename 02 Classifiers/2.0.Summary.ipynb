## Chapter Summary: Classification Algorithms

This chapter explores a variety of machine learning algorithms designed to solve both linear and nonlinear classification problems.

---

### Core Algorithms Comparison

| Algorithm | Key Strengths | Considerations | Feature Scaling | Regularization |
| :--- | :--- | :--- | :--- | :--- |
| **Decision Trees** | Highly interpretable and easy to visualize. | Prone to overfitting if not pruned or used in ensembles. | scale-invariant | |
| **Logistic Regression** | Supports online learning via Stochastic Gradient Descent (SGD) and provides class probabilities. | Primarily a linear classifier. | | |
| **SVMs** | Powerful linear models; can handle nonlinear data using the **kernel trick**. | Requires intensive hyperparameter tuning for optimal performance. | | |
| **Random Forests** | Robust ensemble method; less prone to overfitting than single trees; requires minimal tuning. | Can be less interpretable than a single decision tree. | scale-invariant | |
| **KNN** | Lazy learning approach; requires zero training time. | High computational cost during the prediction phase. | | KNN is a non-parametric method that stores all training data. It does not have parameters (like weights) to constrain through L1/L2 penalties.|
