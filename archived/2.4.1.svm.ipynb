# **1. Core Concept: Maximum Margin Classification**
Support Vector Machines are an extension of the perceptron. While the perceptron focuses on minimizing misclassification errors, SVMs aim to **maximize the margin**.

* **Decision Boundary:** The hyperplane that separates different classes.
* **Margin:** The distance between the decision boundary and the closest training examples from each class.
* **Support Vectors:** The specific training examples that are closest to the decision boundary. These points are the "pillars" that define the location of the hyperplane.


> **Intuition:** A larger margin typically leads to a lower generalization error, whereas a small margin makes the model more prone to overfitting.

---

## 2. Handling Non-linear Data: Slack Variables
In real-world scenarios where data is not perfectly linearly separable, SVMs use **slack variables** to allow for **soft-margin classification**. This introduces a degree of flexibility by allowing some misclassifications to occur during training to achieve a better overall fit.

### The Hyperparameter $C$
The hyperparameter $C$ controls the penalty for misclassification:
* **Large $C$:** High penalty for errors. The model prefers a narrower margin to ensure more points are classified correctly (low bias, high variance/overfitting risk).
* **Small $C$:** Lower penalty for errors. The model allows for a wider margin and is more tolerant of misclassified points (higher bias, low variance).

---

## 3. Comparison: Logistic Regression vs. SVMs

| Feature | Logistic Regression | Support Vector Machines (SVM) |
| :--- | :--- | :--- |
| **Objective** | Maximizes conditional likelihoods. | Maximizes the margin between classes. |
| **Outliers** | More sensitive to outliers. | Robust to outliers (focuses only on support vectors). |
| **Implementation** | Simpler to implement and explain. | More complex math (constrained optimization). |
| **Streaming Data** | Easily updated with new data. | Difficult to update incrementally. |
