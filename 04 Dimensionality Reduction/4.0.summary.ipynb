## Compressing Data via Dimensionality Reduction

**Feature extraction** is a technique used to transform a dataset into a new feature subspace of lower dimensionality, effectively summarizing information content. This process of **data compression** is essential for efficiently storing and analyzing the massive datasets produced in modern technology.

This summarizes fundamental **dimensionality reduction** techniques utilized for **feature extraction**. It highlights methods for projecting data into lower-dimensional spaces to optimize data representation and visualization.

---

### 1. Principal Component Analysis (PCA)

PCA is a fundamental data preprocessing technique that projects data onto a lower-dimensional subspace.

* **Unsupervised nature:** It operates by ignoring the class labels of the dataset.


* **Objective:** The primary goal is to maximize the variance along the orthogonal feature axes.

---

### 2. Linear Discriminant Analysis (LDA)

LDA is a **supervised** dimensionality reduction technique that focuses on maximizing **class separability**.

* **Supervised nature:** It considers class information present in the training dataset. Unlike PCA, LDA uses class labels to find the feature subspace that optimizes the distance between different classes.


* **Objective:** It attempts to maximize the class separability within a linear feature space.It projects the data onto a lower-dimensional space that best segregates the data points based on their respective categories.

---

### 3. Nonlinear Techniques & Visualization

For datasets that are not linearly separable, nonlinear dimensionality reduction methods are utilized.

* **t-SNE:** t-distributed Stochastic Neighbor Embedding is a popular nonlinear technique specifically used for **data visualization**, helping to map high-dimensional data into 2D or 3D spaces.
