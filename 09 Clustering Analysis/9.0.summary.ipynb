## **Working with Unlabeled Data - Clustering Analysis**

Unlike supervised learning—where models are trained on data with known class labels—**cluster analysis** belongs to a category of **unsupervised learning** techniques. This approach allows us to discover hidden structures in data where the correct answers are not known upfront.

The goal of **clustering** is to find a natural grouping in data so that items within the same cluster are more similar to each other than to those from different clusters.

---

### **Key Concepts Coverd**

Clustering is highly exploratory and helps organize data into meaningful structures using three main approaches:

* **k-means algorithm:** Finding centers of similarity.
  * A prototype-based approach that organizes data into spherical clusters based on a predetermined number of centroids. Because it lacks ground-truth labels, its performance is measured using intrinsic metrics like the **elbow method** and **silhouette analysis**.

* **Hierarchical clustering:** Building clustering trees using a bottom-up approach.
  * This approach does not require specifying the number of clusters in advance. Its results can be visualized using a **dendrogram** to help interpret the data.

* **Density-based clustering:** Identifying arbitrary shapes of objects based on data density.
  * A density-based algorithm that groups points according to their local densities. It stands out for its ability to handle outliers and identify **non-globular shapes**.
