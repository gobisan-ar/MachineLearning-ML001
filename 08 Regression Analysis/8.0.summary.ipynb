## Summary: Predicting Continuous Target Variables with Regression Analysis

### Overview

While classification tasks focus on predicting categorical variables or group memberships, **regression analysis** is a subcategory of supervised learning used to predict target variables on a continuous scale.

### Applications

Regression models are highly valuable in both science and industry. They are commonly used for:

* Understanding relationships between different variables.
* Evaluating trends.
* Making forecasts (e.g., predicting a company's future sales).

### Simple Linear Regression

Simple (univariate) linear regression models the relationship between a single feature (explanatory variable, $x$) and a continuous target (response variable, $y$).

* **Equation:** $y = w_1x + b$
* $w_1$: Weight coefficient (slope of the line).
* $b$: Bias unit ($y$-axis intercept).


* **Concept:** The objective is to learn the weights that define the best-fitting straight line (the **regression line**) through the training examples.
* **Residuals:** The vertical distances or errors between the actual training data points and the fitted regression line are called offsets or **residuals**.

### Multiple Linear Regression

Multiple linear regression generalizes the univariate model to handle multiple explanatory variables.

* **Equation:** $y = w_1x_1 + ... + w_mx_m + b = \sum_{i=1}^{m} w_ix_i + b = w^Tx + b$
* **Concept:** Instead of a simple straight line, the fitted model becomes a **hyperplane**. For instance, a model utilizing two features fits a two-dimensional hyperplane within a three-dimensional space.

### Key Takeaways

While visualizing linear regression becomes highly complex with three or more features, the underlying principles remain the same. Simple and multiple linear regression share the core foundational concepts, utilize the same evaluation techniques, and can generally be implemented using compatible code structures.
